## Validation of dataset

```python
test1=test.reset_index()
```

```python
train2 = train1.drop(['index'] , axis=1)
val2 = val1.drop(['index'] , axis=1)
test2 = test1.drop(['index'] , axis=1)
```

## Image size assigning

```python
IMG_SIZE=128
```

```python
df= train2
```

```python
df.head()
```

## Reshaping of image

```python
x_train=[]
y_train=[]
for t in range(len(df)):
    folder = df["Folder"][t]
    file1= df["File"][t]
    image = cv2.imread(os.path.join(folder, file1)) 
    image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
    image = image/255.0
    image = image.reshape(-1,IMG_SIZE,IMG_SIZE,3)
    image = image.reshape(image.shape[1], image.shape[2],image.shape[3])  
    x_train.append(image)
    y_train.append(df["Label"][t])
```

```python
y_train
```

```python
x_train= np.array(x_train)
print("x_train has done")
y_train = np.array(pd.get_dummies(y_train), dtype= np.uint8)
```

```python
x_train = x_train.reshape(-1, IMG_SIZE,IMG_SIZE, 3)
```

```python
df=val2
```

```python
IMG_SIZE
```

```python
x_val=[]
y_val=[]
for t in range(len(df)):
    folder = df["Folder"][t]
    file1= df["File"][t]
    image = cv2.imread(os.path.join(folder, file1)) 
    image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
    image = image/255.0
    image = image.reshape(-1,IMG_SIZE,IMG_SIZE,3)
    
    x_val.append(image)
    y_val.append(df["Label"][t])
    
```

```python
x_val = np.array(x_val)
print("x_val has done")
y_val = np.array(pd.get_dummies(y_val), dtype= np.uint8)
```

```python
x_val.shape
```

```python
x_val = x_val.reshape(-1, IMG_SIZE,IMG_SIZE, 3)
```

## Tensorflow and keras import

```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
from tensorflow.keras.models import Sequential, Model
```

```python
from keras.layers import Dropout, Conv2D, Dense, BatchNormalization, MaxPooling2D, Flatten,Activation, Dense
```

```python
model = Sequential()

model.add(Conv2D(128, 3, 3, input_shape=(x_train.shape[1:])))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
BatchNormalization()
model.add(Dense(512))
model.add(Dropout(0.5))
model.add(Activation('relu'))
BatchNormalization()
model.add(Dense(2))
model.add(Activation('sigmoid'))
```

```python
x_train.shape[1:]
```

```python
model.summary()
```

## Accuracy checking of validation test set

```python
model.compile(loss = 'categorical_crossentropy', optimizer = "adam", metrics = ["accuracy"])
```

```python
hist = model.fit(x_train, y_train, validation_data=(x_val, y_val), epochs = 25)
```

```python
import matplotlib.pyplot as plt

acc = hist.history['accuracy']
val_acc = hist.history['val_accuracy']
loss = hist.history['loss']
val_loss = hist.history['val_loss']

epochs = range(1, len(acc) + 1)

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

```python
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
